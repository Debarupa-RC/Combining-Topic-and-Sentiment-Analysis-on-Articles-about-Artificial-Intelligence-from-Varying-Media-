# Loading and cleaning of the data

```python
#Stats
import pandas as pd

#Support
import csv

#Visualisation
import matplotlib.pyplot as plt
```

```python
# Read the AI_news.csv file and remove the first column since it isn't meaningful
news_ai_df = pd.read_csv("data/AI_news.csv").iloc[:, 1:]
news_ai_df
```

```python
# Remove the 'category' column since we don't use it and it contains many NaN values
news_ai_df = news_ai_df.drop(columns='category')
news_ai_df
```

```python
def remove_empty_rows(df):
    """
    Removes rows that are empty, NaN values,
    or that are unusable (e.g., "No description found").
    """
    
    df = df.dropna()
    df = df[df["description"] != "No description found"]
    df = df[df["body"] != "No content found"]
    df = df[df["description"] != " "]
    df = df[df["body"] != " "]
    
    return df
```

```python
news_ai_df = remove_empty_rows(news_ai_df)
```

```python
# Get year and month
news_ai_df["year"] = news_ai_df["datetime"].str[:4]
news_ai_df["month"] = news_ai_df["datetime"].str[5:7]
```

```python
# Convert the respective columns into strings
news_ai_df['title'] = news_ai_df['title'].astype(str)
news_ai_df['description'] = news_ai_df['description'].astype(str)
news_ai_df['body'] = news_ai_df['body'].astype(str)
news_ai_df['outlet'] = news_ai_df['outlet'].astype(str)
```

```python
news_ai_df.to_csv("data/cleaned_data.csv")
```

### Evaluating the data

```python
# Display the number of articles published per year per outlet in a timeline graph to get an overview
news_ai_df.groupby(["year", "outlet"]).count().iloc[:,0].unstack().fillna(0).plot(title="Number of articles per year")
plt.savefig("plots/articles_per_year_lineplot.png", bbox_inches="tight")
```

```python
# Display a graph of the no. of articles published per year
news_ai_df.groupby("year").count().iloc[:,0].sort_index().plot(kind="bar",
                                                               colormap="summer",
                                                               title="Articles published per year")
plt.savefig("plots/articles_per_year.png", bbox_inches="tight")
```

```python
# Display a graph of the no. of articles published per month
news_ai_df.groupby("month").count().iloc[:,0].sort_index().plot(kind="bar",
                                                                colormap="summer",
                                                                title="Articles published per month")
plt.savefig("plots/articles_per_month.png", bbox_inches="tight")
```

```python
# Display a graph of the no. of articles published per outlet
news_ai_df.groupby("outlet").count().iloc[:,0].sort_values().plot(kind="bar",
                                                                  colormap="summer",
                                                                  title="Articles published per outlet")
plt.savefig("plots/articles_per_outlet.png", bbox_inches="tight")
```
